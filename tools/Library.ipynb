```python
# import부분
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
```

## pandas numpy, pytorch의 차이
- 자료형에 구분이 있어서 이 부분을 잘 모르면 코드가 돌아가지 않음

```python
##데이터 불러오기

from sklearn.datasets import load_iris # 코렙 환경에서는 데이터가 없기 때문에 sklearn을 통해 데이터를 가져옴

iris = load_iris()
```

```python
iris.keys()
```


## numpy의 array 자료형
모델이 받아들이기 편하게 데이터를 구성해서 보여주는 자료형


```python
# iris의 데이터 중 'data'값만 추출

iris_data = iris['data']
print(type(iris_data))
iris_data
```

```python
## pandas dataframe : 사람이 보기 편하게 데이터를 처리 및 가공함
## 모델을 처리할때는 주로 사용하지 않는다.
## 열 이름을 통해 해당 숫자들이 어떤 의미를 갖는지 보여줌

iris_colname = iris['feature_names']
iris_colname
```

```python
iris_pandas = pd.DataFrame(iris_data, columns=iris_colname)
print(type(iris_pandas))
iris_pandas
```

```python
## pytorch
## tensor라는 자료구조를 통해서 데이터를 gpu에 올려서 딥러닝 처리가 가능하게 해줍니다.'
## numpy랑 동일하게 둘다 모델이 처리하기 편하게 데이터를 보여준다.
## 따라서 서로 변환이 쉽다. 사용되는 문법도 굉장히 비슷하다
iris_pytorch = torch.tensor(iris_data)
print(type(iris_pytorch))
iris_pytorch
```

```python
# numpy 와 tensor는 서로 변환이 되기 때문에 쉽게 변환해주는 메소드(함수)를 가지고 있다.
```

```python
iris_pytorch.numpy()
```
