## IOU

```python
def IoU(bbox1, bbox2) :
  # bbox1 = [x1, y1, x2, y2]
  # bbox2 = [x1, y1, x2, y2]
 
  # 교집합 구하기 (좌표)
  intersection_x1 = max(bbox1[0], bbox2[0])
  intersection_y1 = max(bbox1[1], bbox2[1])
  intersection_x2 = min(bbox1[2], bbox2[2])
  intersection_y2 = min(bbox1[3], bbox2[3])

  # 교집합 구하기 (영역)
  intersection = max((intersection_x2 - intersection_x1),0) * max((intersection_y2 - intersection_y1),0) 
  # max 를 쓰는 이유 -> 교집합 부분이 있으면 해당 값이 양수가 나오고  교집합 부분이 없으면 음수가 나오기 때문에 교집합 부분이 없을 경우 0을 만들기 위해서!!
 
  # 합집합 영역
  box1_area = abs(bbox1[2] - bbox1[0]) * abs(bbox1[3] - bbox1[1]) # (x2 - x1) * (y2 - y1)
  box2_area = abs(bbox2[2] - bbox2[0]) * abs(bbox2[3] - bbox2[1]) # (x2 - x1) * (y2 - y1)
  
  unionArea = box1_area + box2_area - intersection
  return intersection / unionArea
```
