```python
import cv2 # opencv import
```

```python
img = cv2.imread('/content/img.jpg') # 이미지를 불러올때 사용하는 함수
```

```python
# cv2.imshow(img) 이미지를 볼때 사용 -> 하지만 colab에서는 호환이 안됨(사용 불가능)

# 다른 방법 사용
from matplotlib import pyplot as plt
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # 이미지를 바꾸는 과정이 필요함
plt.imshow(img)
```

## Bounding Box 생성하기

## 1. 라인 그리기

```python
cv2.line(img, (55 ,55 ), (700, 700), (0,0,255), 5) # 이미지, 시작하는 점의 x, y좌표 / 끝점의 x, y 좌표, [색상,선의 굵기 등등]

# (0,0,255) -> R, G, B에 관련된 색상 값
# 아래 결과 -> 이미지가 가지고 있는 픽셀 값
```

```python
plt.imshow(img)
```

## 2. rectangular 그리기

```python
cv2.rectangle(img, (100 ,100 ), (350, 350), (0,255,0), 5) # 이미지, 시작하는 점의 x, y좌표 / 끝점의 x, y 좌표, [색상,선의 굵기 등등]
```

```python
plt.imshow(img)
```

## 3. bounding box에다가 텍스트 넣기

```python
text = 'Hello world'
img = cv2.putText(img, text, (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0),1) # 이미지, 텍스트, 위치(x,y) 좌표, textStyle(font, )
```

```python
plt.imshow(img)
```
