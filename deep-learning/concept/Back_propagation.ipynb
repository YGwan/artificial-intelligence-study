```python
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
```

## 역전파 - 가중치 업데이트

```python
torch.manual_seed(1017)

inputs = torch.randn(2,2) # 2행 2열짜리 데이터를 생성
inputs
```

```python
torch.manual_seed(777)
targets = torch.randn(2) # label 생성
targets
```

```python
torch.manual_seed(1017)

total_network = nn.Sequential(
    nn.Linear(2,1),
    nn.Sigmoid()
    )

total_network[0].weight
```

```python
# 역전파에 필요한 요소들을 설정
learning_rate = 0.001
criterion = nn.MSELoss(); # LOSS 함수 정의 : MSE
optimizer = torch.optim.Adam(total_network.parameters(), lr = learning_rate)
```

```python
total_network.train()
pred = total_network(inputs)
print(pred)

loss = criterion(targets, pred)
print(loss)
```

```python
loss.backward() # loss값을 참고해서 각 가중치에 대해서 편미분을 하겠다.
optimizer.step() # 편미분된 값과 미리 지정한 학습률을 통해 한 스템 최적화 진행
```

```python
total_network[0].weight

# 가중치(weight) 가 업데이트 된 것을 확인할 수 있다.
```
